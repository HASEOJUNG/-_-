```python
# 데이터 전치
data_transposed = data_jeon.transpose()

# 첫 번째 행을 열 이름으로 설정
data_transposed.columns = data_transposed.iloc[0]

# 첫 번째 행 제거
data_transposed = data_transposed[1:]

# 자치구별로 꺾은선 그래프 그리기
plt.figure(figsize=(16, 10))

for i in range(len(data_transposed.columns)):
    plt.plot(data_transposed.index, data_transposed.iloc[:, i], label=data_transposed.columns[i])

plt.title('서울시 자치구별 주택가격지수(전세) 변화 (2003-2022)')
plt.xlabel('연도')
plt.ylabel('주택가격지수')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 데이터 전치
data_tmae = data_mae.transpose()

# 첫 번째 행을 열 이름으로 설정
data_tmae.columns = data_tmae.iloc[0]

# 첫 번째 행 제거
data_tmae = data_tmae[1:]

# 자치구별로 꺾은선 그래프 그리기
plt.figure(figsize=(16, 10))

for i in range(len(data_tmae.columns)):
    plt.plot(data_tmae.index, data_tmae.iloc[:, i], label=data_tmae.columns[i])

plt.title('서울시 자치구별 주택가격지수(매매) 변화 (2003-2022)')
plt.xlabel('연도')
plt.ylabel('주택가격지수')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

###주택가격지수가 꾸준하게 증가해 오다가 전세와 매매 모두 2020년 부터 꺽이는 추이

```python
# 주택가격지수(매매) 데이터 불러오기
data_sale_path = '/content/gdrive/MyDrive/23-2outlier/주택가격지수(매매).csv'
data_sale = pd.read_csv(data_sale_path, index_col='자치구별')

# 주택가격지수(전세) 데이터 불러오기
data_rent_path = '/content/gdrive/MyDrive/23-2outlier/주택가격지수(전세).csv'
data_rent = pd.read_csv(data_rent_path, index_col='자치구별')

# 주택가격지수(매매)와 주택가격지수(전세)의 차이를 나타내는 데이터프레임 생성
data_diff = data_sale.subtract(data_rent)

# 데이터프레임 확인
data_diff
```

```python
# 데이터프레임을 CSV 파일로 내보내기
data_diff.to_csv('/content/gdrive/MyDrive/23-2outlier/data_diff.csv', index=True)
```

```python
# 연도별로 차이의 평균 계산
mean_difference_by_year = data_diff.mean(axis=0)

# 시각화 (막대 그래프)
plt.figure(figsize=(10, 6))
mean_difference_by_year.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('연도별 주택가격지수(매매) - 주택가격지수(전세) 평균 차이')
plt.xlabel('연도')
plt.ylabel('가격지수 차이')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 연도별로 차이의 평균 계산
mean_difference_by_year = data_diff.mean(axis=0)

# 시각화 (막대 그래프)
plt.figure(figsize=(10, 6))
mean_difference_by_year.plot(kind='bar', color='skyblue', edgecolor='black')

mean_difference_by_year.plot(color='red')

plt.title('연도별 주택가격지수(매매) - 주택가격지수(전세) 평균 차이')
plt.xlabel('연도')
plt.ylabel('가격지수 차이')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
